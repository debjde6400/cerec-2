```python
import matplotlib.pyplot as plt

x = [2,4,6,8,10]
y = [6,7,8,2,4]

x2 = [1,3,5,9,11]
y2 = [7,8,2,4,2]

plt.bar(x,y, label='Bars1', color='r')
plt.bar(x2,y2, label='Bars2', color='c')

plt.xlabel('Soudamini')
plt.ylabel('Shankar')
plt.title('Soudaminir Songsar')

plt.legend()
```

```python
population_ages = [22,55,62,45,21,22,34,42,42,4,99,
                   102,110,120,121,122,130,111,115,112,80,
                   75,65,54,44,83,37,48]

ids = [ x for x in range(len(population_ages))]

plt.bar(ids, population_ages, label='People')

plt.xlabel('Ids')
plt.ylabel('Population')
plt.title('People')

plt.legend()
```

```python
bins = [x for x in range(0, 132, 10)]

plt.hist(population_ages, bins, histtype='bar', rwidth = 0.8)

plt.xlabel('Ids')
plt.ylabel('Population')
plt.title('People')

plt.legend()
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[], color='m', label='Sleeping', linewidth=5)
plt.plot([],[], color='c', label='Eating', linewidth=5)
plt.plot([],[], color='r', label='Working', linewidth=5)
plt.plot([],[], color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])
plt.legend()
```

```python
slices = [7,2,2,13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c','m','r','b']

plt.pie(slices, labels=activities, colors=cols, startangle=90, autopct='%1.1f%%')
```
