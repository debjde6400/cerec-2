```python
pattern_new_sent = dict()
i = 0
applied = 0
non_compliant = []
cnt = len(dff)
fl = open('./resources/output/complience_output.txt', 'w+')

for row in dff.itertuples():
  if (row[2] == 1 and row[4] == row[4]):
    #print(str(row[0]), str(re.split(row[4], row[1], flags=re.IGNORECASE)))
    try:
      sent = SNLP.CoreNLPAnnotator(url='https://d3e3301f6d8e.ngrok.io/').create_sentence(row[1])
    except:
      print('Parse error.')
      non_compliant.append((row[1], row[4]))
      continue
    complied = False

    for p in arr:
      if p.isCompliant(sent):
        g1 = p.causalityExtraction.generateGraphFromSentence(sent)
        p1 = g1.root.origin.phrase[-1]
        p2 = g1.root.target.phrase[-1]

        if p1 is None or p2 is None:
          continue
        if p not in pattern_new_sent:
          pattern_new_sent[p] = [(sent, g1)]
        else:
          pattern_new_sent[p].append((sent, g1))
        applied += 1
        complied = True
        break

    if not complied:
      non_compliant.append((row[1], row[4]))

    print(f'Sentence {i}: done\n')

  i += 1



for k, v in pattern_new_sent.items():
  fl.write(f"{k.getStructure().toString()}: \n")
  for pr in v:
    fl.write(f"{pr[0].getRootConstituent().getCoveredText()}\n")
    fl.write(f"{str(pr[1])}\n")
    
fl.write(f'Out of {len(arr)}, {len(pattern_new_sent)} patterns have compliant sentences.\n')
fl.write(f'Out of {cnt} sentences, complete extraction of cause and effects have been possible for {applied}.\n')
fl.write(f'Ratio between applied and total: {applied/cnt}\n')
fl.write(f'No. of sentences not complying or fully complying with any stored pattern: {len(non_compliant)}\n')
fl.close()

with open("./resources/output/pattern_ns.bin", 'wb') as f:
  pickle.dump(pattern_new_sent, f)
 
with open("./resources/output/non_c.bin", 'wb') as f:
  pickle.dump(non_compliant, f)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
#nitialize our model and tokenizer:
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
model = AutoModel.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
###Tokenize the sentences like before:
sent = [
    "Library management system",
    "Document",
    "Jelly fish",
    "commotion"
]
# initialize dictionary: stores tokenized sentences
token = {'input_ids': [], 'attention_mask': []}
for sentence in sent:
    # encode each sentence, append to dictionary
    new_token = tokenizer.encode_plus(sentence, max_length=128,
                                       truncation=True, padding='max_length',
                                       return_tensors='pt')
    token['input_ids'].append(new_token['input_ids'][0])
    token['attention_mask'].append(new_token['attention_mask'][0])
# reformat list of tensors to single tensor
token['input_ids'] = torch.stack(token['input_ids'])
token['attention_mask'] = torch.stack(token['attention_mask'])
```

```python
output = model(**token)
output.keys()
```

```python
embeddings = output.last_hidden_state
att_mask = token['attention_mask']
mask = att_mask.unsqueeze(-1).expand(embeddings.size()).float()
mask_embeddings = embeddings * mask
summed = torch.sum(mask_embeddings, 1)
summed_mask = torch.clamp(mask.sum(1), min=1e-9)
mean_pooled = summed / summed_mask
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#Let's calculate cosine similarity for sentence 0:
# convert from PyTorch tensor to numpy array
mean_pooled = mean_pooled.detach().numpy()
# calculate
cosine_similarity(
    [mean_pooled[0]],
    mean_pooled[1:]
)
```
